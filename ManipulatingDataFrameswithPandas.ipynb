```python
import pandas as pd
import numpy as np

weather = pd.read_csv('data_folder/pittsburgh2013.csv', index_col= 'Date', parse_dates=True)

# weather.head()

# Write a function to convert degrees Fahrenheit to degrees Celsius: to_celsius
def to_celsius(F):
    return 5/9*(F - 32)

# Apply the function over 'Mean TemperatureF' and 'Mean Dew PointF': df_celsius
df_celsius = weather[['Mean TemperatureF', 'Mean Dew PointF']].apply(to_celsius)


# apply   df.apply(lambda n: n//12) 
#  df.floordiv(12) # Convert to dozens unit

#same slice here
print(weather[['Mean TemperatureF', 'Mean Dew PointF']] is weather.loc[:,['Mean TemperatureF', 'Mean Dew PointF']]) # not same object
print(weather[['Mean TemperatureF', 'Mean Dew PointF']] == weather.loc[:,['Mean TemperatureF', 'Mean Dew PointF']])





# Reassign the column labels of df_celsius
df_celsius.columns = ['Mean TemperatureC', 'Mean Dew PointC']

# Print the output of df_celsius.head()
print(df_celsius.head())
```

```python
import pandas as pd
import numpy as np

election = pd.read_csv('data_folder/pennsylvania2012_turnout.csv', index_col= 'county')

# Create the dictionary: red_vs_blue
red_vs_blue = {'Obama':'blue', 'Romney':'red'}

# Use the dictionary to map the 'winner' column to the new column: election['color']
election['color'] = election.winner.map(red_vs_blue)     # add columns and map the keys and values

# Print the output of election.head()
print(election.head())



# Import zscore from scipy.stats
from scipy.stats import zscore 

# Call zscore with election['turnout'] as input: turnout_zscore
turnout_zscore = zscore(election['turnout'])

# Print the type of turnout_zscore
print(type(turnout_zscore))

# Assign turnout_zscore to a new column: election['turnout_zscore']
election['turnout_zscore'] = turnout_zscore

# Print the output of election.head()
print(election.head())
```

multiple level of a multiindex

```python
s_dict = {'eggs': {0: 47, 1: 110, 2: 221, 3: 77, 5: 205},
 'month': {0: 1, 1: 2, 2: 1, 3: 2, 5: 2},
 'salt': {0: 12.0, 1: 50.0, 2: 89.0, 3: 87.0, 5: 60.0},
 'spam': {0: 17, 1: 31, 2: 72, 3: 20, 5: 55},
 'state': {0: 'CA', 1: 'CA', 2: 'NY', 3: 'NY', 5: 'TX'}}

sales = pd.DataFrame(s_dict)
print(sales) 

# Set the index to be the columns ['state', 'month']: sales
sales = sales.set_index(['state', 'month'])

# Sort the MultiIndex: sales
sales = sales.sort_index()

# Print the sales DataFrame
print(sales)
print('\n')

# Look up data for NY in month 1 in sales: NY_month1
NY_month1 = sales.loc[(['NY'], 1),:]
print(NY_month1)

print('\n')
# Look up data for CA and TX in month 2: CA_TX_month2
CA_TX_month2 = sales.loc[(['CA', 'TX'],  2),:]
print(CA_TX_month2)

print('\n')
# Access the inner month index and look up data for all states in month 2: all_month2
all_month2 = sales.loc[(slice(None),2),:]  #slice both index using slice(None) tuple only
print(all_month2)
```
